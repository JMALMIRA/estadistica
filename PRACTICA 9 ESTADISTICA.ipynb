```python
import pandas as pd
import matplotlib.pyplot as plt

# Crear una figura con 2 filas y 3 columnas
fig, axes = plt.subplots(2, 3, figsize=(15, 8))
axes = axes.flatten()  # Aplanar para acceder fácilmente

# Leer y graficar cada fichero
for i in range(6):
    filename = f"dat{i+1}.csv"  # dat1.csv hasta dat6.csv
    data = pd.read_csv(filename, header=None).values.flatten()  # Leer como vector

    # Dibujar histograma en la subfigura correspondiente
    axes[i].hist(data, bins=10)
    axes[i].set_title(f'Histograma de dat{i+1}.csv')
    axes[i].set_xlabel('Valor')
    axes[i].set_ylabel('Frecuencia')

# Ajustar espaciado
plt.tight_layout()
plt.show()
```

```python
# 3. Comparar distribuciones teóricas con datos generados

# Generar los datos de las distribuciones
d1 = np.random.binomial(200, 0.1, 10000)
d2 = np.random.binomial(200, 0.7, 5000)
d3 = np.random.poisson(200 * 0.1, 10000)
d4 = np.random.poisson(200 * 0.7, 10000)
d5 = np.random.geometric(0.3, 7000)
```

```python
d1 = np.random.binomial(200, 0.1, 10000)
# Histogramas y análisis de parámetros
plt.hist(d1, bins=10, edgecolor='black')
plt.title('Histograma de d1 (Binomial)')
plt.show()
p1 = 1 - np.var(d1) / np.mean(d1)
n1 = np.mean(d1) / p1
print(f"Binomial: p1 = {p1}, n1 = {n1}")
```

```python
plt.hist(d2, bins=10, color='green', edgecolor='black')
plt.title('Histograma de d2 (Binomial)')
plt.show()
p2 = 1 - np.var(d2) / np.mean(d2)
n2 = np.mean(d2) / p2
print(f"Binomial: p2 = {p2}, n2 = {n2}")
```

```python
plt.hist(d3, bins=10, color='red', edgecolor='black')
plt.title('Histograma de d3 (Poisson)')
plt.show()
lambda_ = np.mean(d3)
print(f"Poisson: lambda = {lambda_}")
```

```python
plt.hist(d4, bins=10, color='purple', edgecolor='black')
plt.title('Histograma de d4 (Poisson)')
plt.show()
lambda_ = np.mean(d4)
print(f"Poisson: lambda = {lambda_}")
```

```python
plt.hist(d5, bins=10, color='orange', edgecolor='black')
plt.title('Histograma de d5 (Geométrica)')
plt.show()
p5 = 1 / (1 + np.mean(d5))
print(f"Geométrica: p5 = {p5}")
```
